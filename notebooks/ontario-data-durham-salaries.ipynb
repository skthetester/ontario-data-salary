```python
import pandas as pd
import numpy
import matplotlib.pyplot
import os
```

```python
DATA_DIR = os.path.dirname(os.path.abspath('')) + '\\data\\03_processed\\'
df = pd.read_csv(DATA_DIR + 'processed_durham.csv')
print('Total number of records: ' , df.shape[0])
#df
```

```python
top_10_sector = df.query('Salary_crnt > 100000').groupby('Sector')['LastName'].count() \
    .reset_index(name='count') \
    .sort_values(['count'], ascending=False) \
    .head(10)
top_10_sector
```

```python
top_10_employers = df.query('Salary_crnt > 100000').groupby('Employer')['LastName'].count() \
    .reset_index(name='count') \
    .sort_values(['count'], ascending=False) \
    .head(10)
top_10_employers
```

```python
top_10_salaries = df.query('Salary_crnt > 100000') \
    [['Sector', 'Employer', 'JobTitle', 'Salary_crnt', 'Salary_prev', 'PrcntSalaryDiff']] \
    .sort_values(['Salary_crnt'], ascending=False) \
    .head(10)
top_10_salaries
```

```python
top_10_increases = df.query('Salary_crnt > 100000') \
    [['Sector', 'Employer', 'JobTitle', 'Salary_crnt', 'Salary_prev', 'PrcntSalaryDiff']] \
    .sort_values(['PrcntSalaryDiff'], ascending=False) \
    .head(10)
top_10_increases
```

```python
salary_increased = df.query('PrcntSalaryDiff > 0')
print('Total salaries increased = ' , salary_increased.shape[0] , 
    ' i.e. ' , numpy.round(100*salary_increased.shape[0]/df.shape[0], decimals=2) , '% of total salaries')

salary_decreased = df.query('PrcntSalaryDiff < 0')
print('Total salaries decreased = ' , salary_decreased.shape[0] , 
    ' i.e. ' , numpy.round(100*salary_decreased.shape[0]/df.shape[0], decimals=2) , '% of total salaries')

salary_remained_same = df.query('PrcntSalaryDiff == 0')
print('Total salaries remained same = ' , salary_remained_same.shape[0] , 
    ' i.e. ' , numpy.round(100*salary_remained_same.shape[0]/df.shape[0], decimals=2) , '% of total salaries')
```
